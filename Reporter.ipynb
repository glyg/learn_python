```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

series = pd.read_excel('series-1800-2015_simplified.xlsx').dropna(axis=0, how='all')
```

```python
def simplify_name(name):
    name = name.lower()
    for char in (", (.'"):
        name = name.replace(char, '_')
    name = name.replace(')', '')
    name = name.replace('__', '_')
    return name



series_simple = [simplify_name(name) for name in series['Série']]
series['names'] = series_simple
series = series.set_index('names').T

series.head()
```
